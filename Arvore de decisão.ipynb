### Importando as bibliotecas

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
````

### Carregando o arquivo

````python
credito=pd.read_csv('credito_novo.csv')
credito
````

### Separando as variáveis explicativas da target

````python
y=credito.mau
````

````python
x=credito.drop('mau', axis=1)
````

### Separando entre treino e teste

````python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
````

#### 2) Vamos para o modelo:

1. Defina um objeto com a função da árvore de decisão vista em aula.

5. Calcule a acurácia na base de treinamento

**1. Declaração da árvore de decisão**

````python
clf = DecisionTreeClassifier(random_state=100)
````

**2. Treinando o modelo de árvore criada com os dados de treinamento**

````python
clf = clf.fit(x_train,y_train)
````

**3. Visualização da árvore**

````python
plt.figure(figsize=(25, 10))
plot_tree(clf,
          filled=True,
          class_names=['Aprovado', 'Reprovado'],
          feature_names=x_train.columns.tolist());
````

**4. Visualização da matriz de classificação**

````python
ConfusionMatrixDisplay.from_estimator(clf, x_train,y_train, display_labels=['Aprovados', 'Reprovados']);
plt.show()
````

**5. Acurácia na base de treinamento**

````python
y_chapeu_teste = clf.predict(x_train)
y_chapeu_teste
````

````python
accuracy_score(y_train, y_chapeu_teste)
````
